__Setup__

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
import sunpy.io
import sunpy.map
import glob

# Find sample data
path_to_files = '/home/btremblay/Documents/dir.HelioHackWeek/sampledata/'
filenames_eit_195 = sorted(glob.glob(path_to_files+'eit_l1*'))
filenames_euvi_195_l = sorted(glob.glob(path_to_files+'*eu_L.fts'))
filenames_euvi_195_r = sorted(glob.glob(path_to_files+'*eu_R.fts'))
filenames_test_set = sorted(glob.glob(path_to_files+'EUVI_mask*'))
```

__SoHO/EIT Data at 195 Ã…__

```python
# For this example, we will only look at the first image.
file_nb = 0
eit_195_data_tmp = sunpy.io.fits.read(filenames_eit_195[file_nb])
eit_195_data = eit_195_data_tmp[0].data
eit_195_header = eit_195_data_tmp[0].header

# Plot image
eit_195 = sunpy.map.Map(eit_195_data, eit_195_header)
cmap_eit195 = plt.get_cmap('sohoeit195')
eit_195.peek(vmin=0, vmax=500, cmap=cmap_eit195)
```

__STEREO EUVI Data at 195 Ã…__

```python
# For this example, we will only look at the first image.
file_nb = 0
euvi_195_l_data_tmp = sunpy.io.fits.read(filenames_euvi_195_l[file_nb])
euvi_195_l_data = euvi_195_l_data_tmp[0].data
euvi_195_l_header = euvi_195_l_data_tmp[0].header
euvi_195_r_data_tmp = sunpy.io.fits.read(filenames_euvi_195_r[file_nb])
euvi_195_r_data = euvi_195_r_data_tmp[0].data
euvi_195_r_header = euvi_195_r_data_tmp[0].header

# Plot
euvi_l = sunpy.map.Map(euvi_195_l_data, euvi_195_l_header)
euvi_l.peek(vmin=-0.5, cmap=cmap_eit195)
euvi_r = sunpy.map.Map(euvi_195_r_data, euvi_195_r_header)
euvi_r.peek(vmin=-0.5, cmap=cmap_eit195)
```
